```python
def split_ticket(value):
    if value.isnumeric():
        return pd.Series([None,value])
    else:
        parts = value.split(' ',1)
        return pd.Series(parts if len(parts)==2 else [parts[0],None])
train[["Ticket_Item","Ticket_Number"]]=train["Ticket"].apply(split_ticket)
train
```
